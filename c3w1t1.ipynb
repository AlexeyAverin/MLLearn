```python
# создание DataFrame сразу
import pandas as pd
import numpy as np

# загрузили файл данных
with open("/home/yurist/Рабочий стол/LINUX/COD/PYTHON/MLLearn/checkins.dat", 'r') as fo:
    lines = fo.readlines()
fo.close()

# убрали все пробелы из строки и заменили разделитель ' | ' на ','
#linesCSV = [(' '.join(line.split())).replace('|', ',')  for line in lines]
linesCSV = [(''.join(line.split())).split('|')  for line in lines]

# убрали границу разделитель
linesCSV.pop(1)

# вырезаем строку заголовков
columns = linesCSV.pop(0)

# генерируем словарь данных для создания DataFrame (column:cell)
linesDF = list()
for line in linesCSV:
    lineDict = { column:cell for column, cell in zip(columns, line) }
    linesDF.append(lineDict)
    
# создаем DataFrame
df = pd.DataFrame(linesDF)

# заменяем пустые ячейки в столбцах latitude, longitude на np.nan (NaN) 
df['latitude'].replace('', np.nan, inplace = True)
df['longitude'].replace('', np.nan, inplace = True)
df.head()    
```

```python
#
```

```python
# убираем ненужные данные 
df.dropna(subset=['latitude', 'longitude'], inplace=True)

# создали DataFrame координат
X = df[['latitude', 'longitude']].copy()

# создали короткий X
x = X[0:100000]
x.head()
```

```python
# Кластеризация MeanShift

from sklearn.cluster import MeanShift

clustering = MeanShift(bandwidth = 0.1).fit(x)
#clustering = MeanShift(bandwidth = 0.1).fit_predict(x)
```

```python
# обработка данных кластеризации
# clustering.labels_          - метки кластеров
# clustering.cluster_centers_ - координаты центров

# словарь (имя:колличество)
cNamNum = dict()
for cName in clustering.labels_:
    cNumber = cNamNum.setdefault(cName, 0)
    cNumber += 1
    cNamNum[cName] = cNumber
    
# список имен кластеров при условии >15 (колличество)
namesUpLim = list()

# список координат кластеров при условии >15 (колличество)
pointsUpLim = list()
for cName, cNumber in cNamNum.items():
    if cNumber > 15:
        namesUpLim.append(cName)
        pointsUpLim.append(clustering.cluster_centers_[cName])

for point in pointsUpLim:
    print(point)
```

```python
# Определение банеров
offices = [[33.751277, -118.188740],
           [25.867736, -80.324116],  # 25.7863579 -80.21318522
           [51.503016, -0.075479],   # 51.5091222 -0.12303457  0.04794598833662766
           [52.378894, 4.885084],
           [39.366487, 117.036146],
           [-33.868457, 151.205134]]
dxyList = list()
for office in offices:
    for point in pointsUpLim:
        dx = office[0] - point[0]
        dy = office[1] - point[1]
        dxy = (dx**2 + dy**2)**0.5
        dxyList.append(dxy)
print(min(dxyList))
```
